# Question 1

(a) Create a 2-by-3 array of subplots. Fix k = 3 and instantiate an array N = [5, 10, 50] and a variable
size = 100,000.

```python
from scipy.stats import chi2, gamma, norm

N = [5,10,50]
k = 3
s = 100000

fig, axs = plt.subplots(2, 3)
```

(b) Using a for loop, for each value n in N sample a size × n array of samples from the distribution χ2(k).

```python
results = []
for n in N:
    k = 3
    df = n*k/2
    r = chi2.rvs(df, size=[s,n])
    results.append(r)
print(results[0].shape)
print(results[1].shape)
print(results[2].shape)

results_n5 = results[0]
results_n10 = results[1]
results_n50 = results[2]
```

(c) Compute the sample average along each row (i.e. you should get size sample averages), and plot their histogram in a subplot.

```python
mean_5 = np.mean(results_n5, axis=1)
mean_10 = np.mean(results_n10, axis=1)
mean_50 = np.mean(results_n50, axis=1)

fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs[0,0].hist(mean_5, bins=100, density=True, alpha=0.5, label='n=5')
axs[0,1].hist(mean_10, bins=100, density=True, alpha=0.5, label='n=10')
axs[0,2].hist(mean_50, bins=100, density=True, alpha=0.5, label='n=50')
plt.show()
```

(d) Over the histogram (i.e. in the same subplot), plot the exact density of the distribution of sample aver- ages which is given by the gamma distribution described above (Hint: make sure you input the values of N and k correctly, and check how shape and scale parameters are input in the scipy.stats.gamma documentation). You should get an extremely good agreement between the pdf and the histogram.

```python
a5, loc5, scale5 = gamma.fit(mean_5)
a10, loc10, scale10 = gamma.fit(mean_10)
a50, loc50, scale50 = gamma.fit(mean_50)

pdf_5 = gamma.pdf(mean_5, a5, loc5, scale5)
pdf_10 = gamma.pdf(mean_10, a10, loc10, scale10)
pdf_50 = gamma.pdf(mean_50, a50, loc50, scale50)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs[0,0].hist(mean_5, bins=100, density=True, alpha=0.5, label='n=5')
axs[0,0].scatter(mean_5, pdf_5, s = 1, label='gamma fit')
axs[0,1].hist(mean_10, bins=100, density=True, alpha=0.5, label='n=10')
axs[0,1].scatter(mean_10, pdf_10, s = 1, label='gamma fit')
axs[0,2].hist(mean_50, bins=100, density=True, alpha=0.5, label='n=50')
axs[0,2].scatter(mean_50, pdf_50, s = 1, label='gamma fit')
plt.show()
```

(e) In a separate subplot, display the QQ plot of the sample means versus their exact (Gamma) distri-bution.

```python
import statsmodels.api as sm

fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs[0,0].hist(mean_5, bins=100, density=True, alpha=0.5, label='n=5')
axs[0,0].scatter(mean_5, pdf_5, s = 1, label='gamma fit')
axs[0,1].hist(mean_10, bins=100, density=True, alpha=0.5, label='n=10')
axs[0,1].scatter(mean_10, pdf_10, s = 1, label='gamma fit')
axs[0,2].hist(mean_50, bins=100, density=True, alpha=0.5, label='n=50')
axs[0,2].scatter(mean_50, pdf_50, s = 1, label='gamma fit')
sm.qqplot(mean_5,dist=gamma, line='45', fit=True, ax=axs[1,0])
sm.qqplot(mean_10,dist=gamma, line='45', fit=True, ax=axs[1,1])
sm.qqplot(mean_50,dist=gamma, line='45', fit=True, ax=axs[1,2])
plt.show()
```

```python
from scipy.stats import binom
```

Our H0 is Binom(30,0.1)

```python
dist = binom(120, 0.1)
sc = binom.pmf(20, 120, 0.1)
print(f"The probability under H0 of having observed 20 cases:\n{sc}")
cdf = binom.cdf(19, 120, 0.1)
print(f"The probability under H0 of having observed an outcome at least as extreme as 20 cases:\n{cdf}")
p = 1 - cdf

if p < 0.05:
    print('reject null hypothesis')
else:
    print('fail to reject null hypothesis')

print('It tells us that the epidemic is not as prevelent as we thought')
```

While increasing values of N of a binomial distribution, for larges values of N, the sample mean for 300 classes, represented by the Binomial distribution, is given by the Normal distribution.

You therefore use the CLT to test for the value of p. The variance of Binom (30, p) can be easily computed exactly, so it makes sense to use the Z-test. Your confidence level remains at α = 95% and you now observe a total of 945 cases.
Compute the Z-statistic. For this, make sure you can answer the following:

H0: Binom(9000,0.1)

What is the theoretical mean under H0?

```python
theoretical_mean = 9000*0.1
print(f"The theoretical mean is {theoretical_mean}")
```

What is the observed sample mean?

```python
observed_mean = 945
print(f"The observed mean is {observed_mean}")
```

What is the sample size?

9000

```python
sample_size = 9000
print(f"The sample size is {sample_size}")
```

What is the variance of Binom (30, p)?

```python
var = binom.var(n=sample_size, p=0.1)
print(f"The variance is {var}")
```

```python
z = (observed_mean - theoretical_mean)/(np.sqrt(var))
print(f"The z-score is {z}")
```

```python
1 - norm.cdf(z)
print(f"The p value is {1 - norm.cdf(z)}")
```

```python
if p == 0.025: #Assuming a two-tailed test, our rejection region is 0.025 on each side.
    print('reject null hypothesis')
else:
    print('fail to reject null hypothesis')

print('It tells us that the epidemic more prevelent than we thought, compared with the example of n=120')
```
